```python
import numpy as np
import cv2 as cv
import math 
from skimage import io, restoration,data
from matplotlib import pyplot as plt
from PIL import Image
from scipy import ndimage
```

```python
original = data.camera()



levels = int(input("enter:"))
div = 255/(levels - 1)
new_intensity = np.round(original/div)*div




fig,arr1= plt.subplots(1,2)

arr1[0].imshow(original)
arr1[0].set_title('original')

arr1[1].imshow(new_intensity)
arr1[1].set_title('new_intensity')


```

```python
image1 = data.camera()

fig,arr1= plt.subplots(1,2)
fig,arr2= plt.subplots(1,2)


arr1[0].imshow(image1)
arr1[0].set_title('original')



arr1[1].imshow(cv.blur(image1,(3,3)))
arr1[1].set_title('3x3 kernal')

arr2[0].imshow(cv.blur(image1,(10,10)))
arr2[0].set_title('10x10 kernal')



arr2[1].imshow(cv.blur(image1,(20,20)));
arr2[1].set_title('20x20 kernal')
```

```python
im_np = data.camera()

fig,arr1= plt.subplots(1,3)

im = Image.fromarray(im_np)

im_rotated_45 = im.rotate(45, expand=True)
im_rotated_90 = im.rotate(90, expand=True)



arr1[0].imshow(im)
arr1[0].set_title('original')



arr1[1].imshow(im_rotated_45)
arr1[1].set_title('Rotated by 45')

arr1[2].imshow(im_rotated_90)
arr1[2].set_title('Rotated by 90')


```

```python

im = data.camera()

plt.gray()

STRIDE = (3,5,7)
rows,cols = im.shape

f, axarr = plt.subplots(1, len(STRIDE))
for i,S in enumerate(STRIDE):
    im_mod = im.copy()
    for r in range(int(S/2),rows,int(S)):
        for c in range(int(S/2),cols,int(S)):
            im_mod[int(r-S/2):int(r+S/2+1),int(c-S/2):int(c+S/2+1)] = np.average(im[int(r-S/2):int(r+S/2+1),int(c-S/2):int(c+S/2+1)])

    axarr[i].imshow(im_mod);
    axarr[i].set_title('{}x{} average block'.format(S,S))


```
